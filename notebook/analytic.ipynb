```python
import pandas as pd
import numpy as np
import scipy
import glob
import matplotlib.pyplot as plt
```

```python
start_date = "01_01_2019"
end_date = "02_12_2019"
data_path = "../dataProcessing/Data/ext/{}_{}".format(start_date,end_date)
```

```python
file_list = glob.glob(data_path+"/*")
file_list
```

```python
for file in file_list:
    if "UPCOM" in file:
        upcom = pd.read_excel(file)
    elif "Index" in file:
        index = pd.read_excel(file)
    elif "HOSE" in file:
        hose = pd.read_excel(file)
    elif "HNX" in file:
        hnx = pd.read_excel(file)
```

```python
index.tail()
```

```python
df = index[(index["VOLUME"]>0) & (index['TICKER']=='VNINDEX')]
# df = index[(index['TICKER']=='VNINDEX')]
df.tail()
```

```python
x = df['DATE']
y = df['VOLUME']

fig= plt.figure(figsize=(20,10))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

plt.show()
```

```python
x = df['DATE']
i_open = df['OPEN']
close = df['CLOSE']

fig= plt.figure(figsize=(20,10))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,i_open,marker='', color='blue', linewidth=2, linestyle='dashed', label="open")
axes.plot(x,close,marker='', color='olive', linewidth=2)
# show title
plt.legend()
plt.show()
```
